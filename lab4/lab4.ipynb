```
from bs4 import BeautifulSoup as bs
from openpyxl import load_workbook, Workbook
import requests 
import string
import random

wb = Workbook()

file="chmielewski175IC.xlsx"

#Utworzenie arkuszy
ws1=wb.create_sheet("Giełda")
ws2=wb.create_sheet("Linki")
ws3=wb.create_sheet("Filmweb")

wb.active
ws1['A1'] = 'Kod'  
ws1['B1'] = 'Kurs' 
ws1['C1'] = 'Zmiana'  
ws1['D1'] = 'Transakcje'

#Generacja kodów
kurs_out=0
kurs=0
zmiana=0
#transakcja=0
zmiana_out=0
transakcje=0
y=0
codes = []
url = "https://stooq.pl/q/?s="
while len(codes)<=4:
  code = ''.join([random.choice(string.ascii_lowercase) for _ in range(3)])
  
  request = requests.get(url+code)
  soup = bs(request.text, "html.parser")
  error=0
  if "nie istnieje" not in soup.get_text():
    try:
      try:
        kurs = soup.find("span", id=f"aq_{code}_c2")
        kurs_out = float(kurs.text)
        print(kurs)
      except:
        kurs = soup.find("span", id=f"aq_{code}_c1")
        kurs_out = float(kurs.text)
        print(kurs)
      
      zmiana = soup.find("span", id=f"aq_{code}_m3")
      zmiana_out = zmiana.text
      print(zmiana)
      print(code)
      transakcja =soup.find("span", id=f"aq_{code}_n1")
      transakcje = float(transakcja.text)
      print(transakcja)
    except:
      print("Invalid data")
      error=1
      
    if error==1:
      continue
    else:
      codes.append(code)
    # Ustawianie rekordów
    row=y+2
    ws1["A"+ f"{row}"]=codes[y]
    
    ws1["B"+f"{row}"] = kurs_out
    
    ws1["C"+f"{row}"] = zmiana_out
    
    ws1["D"+f"{row}"] = transakcje
    y+=1
#Wypisywanie kodów
for x in range(len(codes)):
  print(codes[x])
#zapis do pliku
wb.save(file)

#ustawianie połączenia
url = "https://zacniewski.gitlab.io"
request = requests.get(url)
soup = bs(request.text, "html.parser")

link_counter = 1

#Funkcja przypisywania linków do rekordów
for a in soup.find_all('a', href=True):
  ws2.cell(row=link_counter, column=1).hyperlink = a['href']
  link_counter = link_counter + 1

wb.save(file)

#ustawianie połączenia
url = "https://www.filmweb.pl/film/Warcraft%3A+Pocz%C4%85tek-2016-334999"
request = requests.get(url)
soup = bs(request.text,"html.parser")

#Nazywanie kolumn
ws3['A1'] = 'Reżyser'
ws3['A2'] = 'Data Premiery'
ws3['A3'] = 'BoxOffice'
ws3['A4'] = 'Ocena filmu'

#Przypisywanie komórek
rezyser = soup.find(itemprop="director").text
ws3['B1'] = rezyser

data_premiery = soup.find("span", {"class": "block"}).text
ws3['B2'] = data_premiery

boxoffice = soup.find('div', attrs={'class':'boxoffice'}).previous_element
ws3['B3'] = boxoffice

ocena = soup.find('span', attrs={'class':'filmRating__rateValue'}).text
ws3['B4'] = ocena

wb.save(file)
```
